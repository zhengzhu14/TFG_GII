## Multistart optimizacion

```python
k = 15
np.random.seed = seed
```

```python
angulos = []
iteraciones = []
for i in range(k):
    x = list(np.random.uniform(0, np.pi, 1)) + list(np.random.uniform(0, 2*np.pi, 1))
    print(x)
    mon, result = qaoa.qaoa_algorithm(circuit, Hc, x, min_method = 'COBYLA')
    angulos.append(list(result.values()))
    iteraciones.append(mon.iterations)
```

```python
evaluaciones = qaoa.evaluate_params(circuit, Hc, angulos)
for i, (e , a) in enumerate(zip(evaluaciones, angulos)):
    print(f'{e:.4f} angulos = {a}, iteraciones = {iteraciones[i]}')
```

## Normalizando el Hamiltoniano

```python
def normalize_hamiltonian (Hc):
    nHc = Hc.copy()
    norma = np.max(np.abs(nHc.coeffs))

    nHc.coeffs = nHc.coeffs / norma

    return nHc
```

```python
nHc = normalize_hamiltonian(Hc)
print(nHc)
```

```python
ncircuit = qaoa.construct_circuit(nHc, reps = 1)
```

```python
nmon, nresult = qaoa.qaoa_algorithm(ncircuit, nHc)

for i, evals in enumerate(nmon.evaluation):
    print(f'{i} Evaluacion = {evals} \tAngulos = {nmon.parameters[i]}')
```

```python
print(nresult)
```

```python
ncmon, ncresult = qaoa.qaoa_algorithm(ncircuit, nHc, min_method = 'COBYLA')
for i, evals in enumerate(ncmon.evaluation):
    print(f'{i} Evaluacion = {evals} \tAngulos = {ncmon.parameters[i]}')
```
